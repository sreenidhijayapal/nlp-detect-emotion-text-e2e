# **Import Libraries**

```python
#import relevant libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# **Data Pre-processing**

```python
#data pre-processing
evcs_dataset = pd.read_csv('error_type_evcs.csv', header = None, encoding='utf8')
#train apriori model
#reshape the format of dataset suitable for apriori function - list of error types
list_of_error_types = []
for i in range(0, 4):
  list_of_error_types.append([str(evcs_dataset.values[i, j]) for j in range(0, 4)]) #loop through every row, deals with NaN values, only string format inputs to apriori, encoding='cp1252'
```

# **Train Apriori**

```python
from apyori import apriori
#return rules for the dataset
#parameters - error types, strongest rule of error association minimum number of times, error type association minimum score, relevance of a rule
rules = apriori(transactions=list_of_error_types, min_support=0.125, min_confidence=0.2, min_lift=3, min_length=2, max_length=4)
```

# **Display Results**

```python
#store results of the rules
results = list(rules)
results
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidence  = [result[2][0][2] for result in results]
    lifts  = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidence, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['First Error Message', 'Following Error Type', 'Support', 'Confidence', 'Lift'])
```
